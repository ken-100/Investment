```python
from xbbg import blp
def BDP(T):

    F = ['etf_undl_index_ticker','crncy','name_kanji','fund_management_co','fund_total_assets','fund_assets_usd','fund_expense_ratio','fund_total_exp']
    df = blp.bdp(tickers=T, flds=F).loc[T,:].reset_index(drop=False)
    df.rename(columns={'index': 'ticker'}, inplace=True)

    df = df.sort_values(by='fund_assets_usd', ascending=False).reset_index(drop=True)
    tmp = ['fund_total_assets','fund_assets_usd']
    df[tmp] = df[tmp].applymap('{:,.0f}'.format)
    
    tmp = list(dict.fromkeys(df['etf_undl_index_ticker'].tolist()))
    T = [x + ' Index' for x in tmp]
    BM = blp.bdp(tickers=T, flds=['name_kanji','long_comp_name']).loc[T,:].reset_index(drop=False)
    BM.rename(columns={'index': 'ticker'}, inplace=True)

    return df, BM
```
