```python
import pdblp
from xbbg import blp
import pandas as pd
import workdays
import datetime
import numpy as np

C = ["US","DM","UK","CD","AD","NZ","SZ","NK","SK"]
T0 = ["L","YC"]
tmp = ["Currency","country_iso","px_last","volatility_90d","rsi_30d"]

T_List = {}
BDP_List = {}

for i in T0:
    T = ["SB" + x + "13" + i + " Index" for x in C]
    BDP0 = blp.bdp(tickers=T, flds=tmp)
    BDP0 = BDP0.loc[T,:]

    for j in tmp[2:]:
        BDP0[j.lower()] = round(BDP0[j.lower()].astype(float),2)
    
    BDP0.insert(1, "Ticker", T)
    BDP0 = BDP0.reset_index(drop=True)
    
    T_List[i] = T
    BDP_List[i] = pd.DataFrame(BDP0.values, columns = BDP0.columns)

BDP_List["YC"]["currency"] = BDP_List["L"]["currency"]
BDP_List["YC"]["country_iso"] = BDP_List["L"]["country_iso"]

BDP_List["L"]
```
