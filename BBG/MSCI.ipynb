```python
data = [
    ['MXWD', 'M2WD', 'M1WD', '-', '-', '-'],
    ['MXWO', 'M2WO', 'M1WO', '-', '-', '-'],
    ['MXEF', 'M2EF', 'M1EF', '-', '-', '-'],
    ['MXEU', 'M2EU', 'M1EU', '-', '-', '-'],
    ['MXUS', 'M2US', 'M1US', 'ES', 'SPX', 'SPXT'],
    ['MXJP', 'M8JP', 'M7JP', 'TP', 'TPX', 'TPXDDVD'],
    ['MXGB', 'M8GB', 'M7GB', 'Z', 'UKX', 'TUKXG'],
    ['MXCA', 'M8CA', 'M7CA', 'PT', 'SPTSX60', '-'],
    ['MXFR', 'M8FR', 'M7FR', 'CF', 'CAC', '-'],
    ['MXCN', 'M8CN', 'M7CN', 'IFB', 'SHSN300', '-'],
    ['MXCH', 'M8CH', 'M7CH', 'SM', 'SMI', '-'],
    ['MXDE', 'M8DE', 'M7DE', 'GX', 'DAX', '-'],
    ['TAMSCI', 'M8TW', 'M7TW', 'FT', 'TWSE', '-'],
    ['MXIN', 'M8IN', 'M7IN', 'NZ', 'NIFTY', '-'],
    ['MXAU', 'M8AU', 'M7AU', 'XP', 'AS51', '-'],
    ['MXNL', 'M8NL', 'M7NL', 'EO', 'AEX', '-'],
    ['MXKR', 'M8KR', 'M7KR', 'KM', 'KOSPI2', '-'],
    ['MXIE', 'M8IE', 'M7IE', '-', 'ISEQ', '-'],
    ['MXDK', 'M8DK', 'M7DK', 'OMW', 'OMXC25', '-'],
    ['MXSE', 'M8SE', 'M7SE', 'QC', 'OMX', '-'],
    ['MXHK', 'M8HK', 'M7HK', 'HI', 'HSI', '-'],
    ['MXES', 'M8ES', 'M7ES', 'IB', 'IBEX', '-'],
    ['MXIT', 'M8IT', 'M7IT', 'SW', 'FTSEMIB', '-'],
    ['MXBR', 'M2BR', 'M1BR', 'BZ', 'IBOV', '-']
]

df = pd.DataFrame(data, columns=['Price', 'Gross', 'Net', 'Future', 'Index', 'Index_Gross'])
tmp = [x + ' Index' for x in df['Price']]
Name = blp.bdp([x + ' Index' for x in df['Price']],['country_iso','name']).loc[tmp,:].reset_index(drop=True)

List = pd.concat([Name, df], axis=1)
List = List.rename(columns={'country_iso':'ISO'})
List
```

```python
ISO = blp.bds('ACWI US Equity','LU_COUNTRY_ISO').reset_index(drop=True) 
ISO.columns = ['ISO','Name']

BBG = blp.bds('ACWI US Equity','LU_CDR_COUNTRY_CODE').reset_index(drop=True)
BBG.columns = ['BBG','Name']
BBG['Name'] = BBG['Name'].str.upper()

Mapping = BBG.set_index('Name')['BBG'].to_dict()
ISO.insert(1, 'BBG',0)
ISO['BBG'] = ISO['Name'].map(Mapping).fillna(ISO['ISO'])

W = blp.bds('ACWI US Equity','HB_GEO_CNTRY_ALLOC')
W.columns = ['Name', 'Weight']
W = W.sort_values(by='Weight', ascending=False).reset_index(drop=True)
W.insert(0, 'ISO',0)
W.insert(1, 'BBG',0)
tmp = W['Name']
W['Name'] = W['Name'].str.upper()

Mapping = ISO.set_index('Name')['ISO'].to_dict()
Mapping.update({'U.S.': 'US', 'U.K.': 'GB','Czech Republic': 'CZ','U.A.E.': 'AE'} )
W['ISO'] = W['Name'].map(Mapping).fillna(W['ISO'])

Mapping = ISO.set_index('Name')['BBG'].to_dict()
Mapping.update({'U.S.': 'US', 'U.K.': 'GB'} )
W['BBG'] = W['Name'].map(Mapping).fillna(W['BBG'])

W['Name'] = tmp
W['Weight']  = round(W['Weight'],2)
W['Cumulative'] = W['Weight'].cumsum()

Mapping = List.set_index('ISO')['Price'].to_dict()
W['MSCI'] = W['ISO'].map(Mapping)
Mapping = List.set_index('ISO')['Index'].to_dict()
W['Index'] = W['ISO'].map(Mapping)
W.loc[W['ISO']=='CN',['MSCI','Index']] = ['M7CN','XIN9I']
W.loc[:19,:]
```

```python
d_from = '20070101'
d_from = '20150101'
d_to = '20240331'
W = W[:20]

T = W['MSCI'].tolist() + W['Index'].tolist()

T1 = ['KOSPI2','NIFTY']
T2 = ['TWSE']
for i in T1 + T2:
    T.remove(i)

df = blp.bdh([t + ' Index' for t in T], 'tot_return_index_gross_dvds', d_from, d_to, Per='M').reset_index() 
df.columns = ['Date'] + T


tmp = blp.bdh([t + ' Index' for t in T1], 'tot_return_index_gross_dvds', d_from, d_to, Days='A', Fill='P', Per='M').reset_index()
tmp = tmp.iloc[:,1:]
tmp.columns= T1
df = pd.concat([df.head(),tmp.head()],axis=1)


tmp = blp.bdh([t + ' Index' for t in T2], 'tot_return_index_gross_dvds', d_from, d_to, Days='A', Fill='P', Per='M').reset_index()
tmp = tmp.iloc[:,1:]
tmp.columns= T2
df = pd.concat([df.head(),tmp.head()],axis=1)


df.iloc[:,1:] = df.iloc[:,1:].pct_change()
df.head()
```
