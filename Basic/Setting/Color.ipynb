```python
import matplotlib.pyplot as plt
import seaborn as sns
import matplotlib.colors as mcolors
import pandas as pd
import matplotlib.colors as mcolors
# https://pythondatascience.plavox.info/wp-content/uploads/2016/06/colorpalette.png

# Use seaborn's default color palette
colors = sns.color_palette()
fig, ax = plt.subplots(1, 1, figsize=(10, 2))
for i, color in enumerate(colors):
    ax.bar(i, 1, color=color)

ax.get_yaxis().set_visible(False)
ax.set_xticks(range(len(colors)))
ax.set_xticklabels([f"color {i}" for i in range(len(colors))])

plt.show()
```

```python
import matplotlib.pyplot as plt

colors = ['blue', 'red', 'green', 'purple']


fig, axs = plt.subplots(3, 2, figsize=(8, 10))

for i, l in enumerate([1, 0.5]):
    for j, a in enumerate([1, 0.5, 0.3]):
        for y, color in enumerate(colors):
            axs[j, i].hlines(y, -3, 3, color=color, alpha=a, linewidth=l)
        axs[j, i].set_title("alpha="+str(a)+", linewidth="+str(l))

plt.tight_layout()
plt.show()
```
