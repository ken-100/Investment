```python
import holidays
import pandas as pd
from datetime import datetime
# https://pypi.org/project/holidays/
```

```python
def get_holidays(country, year=None):
    
    tmp = getattr(holidays, country)(years=year)
    df = [{'date': date, 'name': name} for date, name in sorted(tmp.items())]
        
    return pd.DataFrame(df)


for c in ['US','Japan']:
    display(get_holidays(c, 2025))
```

```python
def get_holidays(country, d_from, d_to):   
    d_from = pd.to_datetime(d_from)
    d_to = pd.to_datetime(d_to)
    
    tmp = list(range(d_from.year, d_to.year + 1))   
    tmp = getattr(holidays, country)(years=tmp)    
    df = [{'date': date, 'name': name} for date, name in sorted(tmp.items())]
    
    df = pd.DataFrame(df)   
    df['date'] = pd.to_datetime(df['date'])
    
    tmp = (df['date'] >= d_from) & (df['date'] <= d_to)
    df = df.loc[tmp].reset_index(drop=True)
    
    return df


d_from = '2023-01-01'
d_to = '2024-09-21'

for c in  ['US', 'Japan']:
    print(c)
    df = get_holidays(c, d_from, d_to)
    display(df)
```
