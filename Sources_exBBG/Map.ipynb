```python
# 東京駅の緯度経度
tokyo_station = [35.6812, 139.7671]

# 地図の作成（スタイルを指定可能）
mymap = folium.Map(
    location=tokyo_station,
    zoom_start=15,
    tiles='CartoDB positron'  # 地図のスタイル
)

# 東京駅のマーカーを追加
folium.Marker(
    tokyo_station,
    popup="Tokyo Station",
    tooltip="Click for more info",
    icon=folium.Icon(color="red", icon="train", prefix='fa')
).add_to(mymap)

# 周辺の主要スポットを追加
landmarks = {
    "丸の内ビル": [35.6817, 139.7633],
    "KITTE": [35.6795, 139.7646],
    "大手町": [35.6855, 139.7677]
}

for name, coords in landmarks.items():
    folium.Marker(
        coords,
        popup=name,
        icon=folium.Icon(color="blue", icon="info-sign")
    ).add_to(mymap)

# 円を追加（半径500m）
folium.Circle(
    tokyo_station,
    radius=500,
    color="red",
    fill=True,
    popup="500m radius"
).add_to(mymap)

# 地図の表示
display(mymap)

# 地図をHTMLファイルとして保存（オプション）
mymap.save("tokyo_station_map.html")
```
