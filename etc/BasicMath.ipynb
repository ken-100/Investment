```python
print(np.exp(2))
print(math.exp(2))
```

```python
print(np.exp(0))
print(np.exp(-2))
# 1.0
# 0.1353352832366127
```

```python
x = list(range(-10,5))
y = np.exp(x)   # Impossible with "math"

fig, ax = plt.subplots(1, 1, squeeze=False,figsize=(8,3),tight_layout=True)
ax[0,0].plot(x, y)
ax[0,0].set_title("y=exp(x)")
plt.show()
```
